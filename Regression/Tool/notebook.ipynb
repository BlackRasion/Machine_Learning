```python
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Load up the X and y data.
For supervised learning, we need a named 'y' target.

```python
X,y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
#select a portion of this dataset to plot
X1 = X[:,2]
X1 = X1.reshape(-1,1) #reshape to 2D array using reshape(n_rows, n_columns). 
                      #If one of the parameter is -1, the corresponding dimension is calculated automatically.
print("X1.shape: ", X1.shape)  
```

```python
#split both the data (X) and the target (y) into test and training sets.
X_train, X_test, y_train, y_test = model_selection.train_test_split(X1,y,test_size=0.33)
```

```python
#Load up the linear regression model and train it with X and y training sets
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
#create a prediction using test data
y_pred = model.predict(X_test)
```

```python
#show data in a plot
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
